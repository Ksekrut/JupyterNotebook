## Прапор Йорданії




```python
def is_prime(n):
    """
    Функція перевіряє, чи є число простим.
    """
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

def task_6(a, b):
    """
    Функція знаходить всі прості числа на відрізку [a, b].
    """
    if a > b:
        a, b = b, a  # Якщо a більше b, міняємо їх місцями

    # Пошук простих чисел на відрізку [a, b]
    primes = [num for num in range(a, b + 1) if is_prime(num)]
    print(primes)


a = 3 
b = 15  
task_6(a, b)

```
